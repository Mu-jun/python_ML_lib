```python
import pandas as pd
```

```python
file_path = './dataset/auto-mpg.csv'
df = pd.read_csv(file_path, header=None)
print(df.head())

# 열 이름 지정
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
             'acceleration','model year','origin','name']
# 데이터 미리보기
# 처음 행부터 5개 df.head(3) -> 3개
print(df.head()); print()
# 마지막 행부터 5개 df.tail(3) -> 3개
print(df.tail()); print()
print('행과 열의 갯수 :',df.shape) # 행과 열의 갯수
```

```python
# 데이터 통계 요약 정보
print(df.describe()); print() # 산술(숫자) 데이터만
print(df.describe(include='all')); print() # NaN 포함
print('각 열의 유효한 자료의 갯수')
print(df.count())
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')

# 각 열의 유효한 자료의 객수
print(titanic.count()); print()

df_t = titanic.copy()
df_t.drop(columns='deck',axis=1,inplace=True)
print(df_t.count())
df_t.info()
```

```python
# 열의 고유한 값의 갯수 : value_counts()
print(df_t['class'].value_counts())
df_t.value_counts()
```

```python
# 데이터프레임, 시리즈 모두 사용 가능
# 평균값 : mean()
print(df.mean()); print()
print('mpg 평균값 :', df.mpg.mean()); print()
# 중간값 : median()
```

```python
# 데이터프레임, 시리즈 모두 사용 가능
# 평균값 : mean()
print(df.mean(numeric_only=True)); print()
print('mpg 평균값 :', df.mpg.mean()); print()
```

```python
# 중간값 : median()
print(df.median(numeric_only=True)); print()
print('mpg 중간값 :', df.mpg.median()); print()
```

```python
# 최대값 : max()
print(df.max(numeric_only=True)); print()
print('mpg 최대값 :', df.mpg.max()); print()
```

```python
# 최소값 : min()
print(df.min(numeric_only=True)); print()
print('mpg 최소값 :', df.mpg.min()); print()
```

```python
# 표준편차 : std()
print(df.std(numeric_only=True)); print()
print('mpg 표준편차 :', df.mpg.std()); print()
```

```python
# 분산 : var()
print(df.var(numeric_only=True)); print()
print('mpg 분산 :', df.mpg.var()); print()
```

```python
# 상관계수 : corr() -> 열 이름 리스트
print('모든 상관계수')
print(df.corr())
print()
print('mpg와 weight의 상관계수')
print(df[['weight','mpg']].corr())
```

```python
file_path = './dataset/auto-mpg.csv'
df = pd.read_csv(file_path, header=None)
print(df.head())

# 열 이름 지정
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
             'acceleration','model year','origin','name']
print(df.count())
```

```python
# 두 변수간의 상관관계를 그림으로 표현
df.plot(kind='scatter',x='weight',y='mpg')

df.plot(kind='hexbin',x='weight',y='mpg')
```

```python
# 모든 데이터 상관관계
ndf = df.loc[:,'mpg':'acceleration']
pd.plotting.scatter_matrix(ndf,figsize=(20,20))
```
